```python
data['Continent'].fillna(data['Continent'].mode()[0], inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data['WHO Region'].fillna(data['WHO Region'].mode()[0], inplace=True)
print(data.isnull().sum())
```

```python

data.loc[data['Country/Region'] == 'Holy See', 'Population'] == data[data['Country/Region'] == 'Italy']['Population'].values[0]
print(data.isnull().sum())
```

```python
data['NewCases'].fillna(0, inplace=True)
data['NewDeaths'].fillna(0, inplace=True)
```

```python
median_deaths = data['TotalDeaths'].median()
data['TotalDeaths'].fillna(median_deaths, inplace=True)

median_deaths_per_million = data['Deaths/1M pop'].median()
data['Deaths/1M pop'].fillna(median_deaths_per_million, inplace=True)
print(data.isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression

train = data.dropna(subset=['TotalRecovered'])
x_train = train[['TotalCases', 'TotalDeaths']]
y_train = train['TotalRecovered']

model = LinearRegression()
model.fit(x_train, y_train)

missing = data[data['TotalRecovered'].isna()]
x_missing = missing[['TotalCases', 'TotalDeaths']]
data.loc[data['TotalRecovered'].isna(), 'TotalRecovered'] = model.predict(x_missing)


print(data.isnull().sum())
```

```python
data['ActiveCases'] = data['TotalCases'] - data['TotalDeaths'] - data['TotalRecovered']
```

```python
median_critical = data['Serious,Critical'].median()
data['Serious,Critical'].fillna(median_critical, inplace=True)
```

```python
data['Tot Cases/1M pop'] = (data['TotalCases'] / data['Population']) * 1e6
print(data.isnull().sum())
```

```python
from sklearn.impute import KNNImputer

train = data.dropna(subset=['TotalTests'])
x_train = train[['Population']]
y_train = train['TotalTests']

imputer = KNNImputer(n_neighbors=5)

data['TotalTests'] = imputer.fit_transform(data[['TotalTests']])
print(data.isnull().sum())



```

```python
from sklearn.impute import KNNImputer

train = data.dropna(subset=['Tests/1M pop'])
x_train = train[['Population']]
y_train = train['Tests/1M pop']

imputer = KNNImputer(n_neighbors=5)

data['Tests/1M pop'] = imputer.fit_transform(data[['Tests/1M pop']])
print(data.isnull().sum())
```

```python
data['Population'].fillna(data['Population'].mean(), inplace=True)
data['Tot Cases/1M pop'].fillna(data['Tot Cases/1M pop'].mean(), inplace=True)
data['NewRecovered'].fillna(data['NewRecovered'].mean(), inplace=True)
print(data.isnull().sum())
```
